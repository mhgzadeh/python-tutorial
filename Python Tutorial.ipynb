6- Lists

they are used to store multiple items which are ordered, changeable, and allow duplicate values

```python
city = ['stuttgart', 'berlin', 'hamburg', 'Munich', 'frankfurt', 'leipzig']
print(f"first city ==> {city[0]}")
print(f"last city ==> {city[-1]}")
print(f"last letter of third city ==> {city[4][-1]}")

city[1] = 'Bonn' # change berlin to bonn
print(f"list of cities ==> {city}")
```

6-1- List builtin methods

append(): add an item to the end of the list
insert(): insert a list item at a specified index
pop(): removes the last item if do not specify index
remove(): removes the specified item
extend(): combine two list with each other
sort() and reverse(): sort list in numeric/alphabetical order or vice versa

```python
city_1 = ['stuttgart', 'berlin']
city_2 = ['Munich', 'frankfurt', 'leipzig']

city_1.append(city_2[0]) # append first city of city_2 to city_1
print(f"append first city of city_2 to city_1 ==> {city_1}")

city_1.insert(1, city_2[1]) # insert second city of city_2 into city_1 as a second city
print(f"insert second city of city_2 into city_1 as a second city ==> {city_1}")

print(f"last city of city_1 is popped ==> {city_1.pop()}")

city_1.remove('berlin') # remove berlin from city_1
print(f"remove berlin from city_1 ==> {city_1}")

city_1.extend(city_2)
print(f"extend city_1 and city_2 ==> {city_1}")

city_1.sort() # 
print(f"sort city_1 ==> {city_1}")

city_1.reverse() # 
print(f"sort reversely city_1 ==> {city_1}")
```

8- Set

Sets are unordered, unchangeable, and do not allow duplicate values.

```python
age_list = [25, 50, 25, 50, 50, 25, 50, 16, 11, 16, 11, 11]
age_set = {25, 50, 25, 50, 50, 25, 50, 16, 11, 16, 11, 11}
print(f"length of age_list is {len(age_list)}, length of age_set is {len(age_set)}")

age_set.add(100)
print(f"add 100 to age_set ==> {age_set}")
age_set.add(200)
print(f"add 200 to age_set ==> {age_set}")
age_set.add(300)
print(f"add 300 to age_set ==> {age_set}") # so it is unordered
```

11- For Loops

A for loop is used for iterating over a sequence (list, tuple, dictionary, set, or string)

```python
years = (2001, 2005, 1988, 1996, 2022)
for year in years:
    print(year)
```

```python
years = (2001, 2005, 1988, 1996, 2022)
for i in range(len(years)):
    print(years[i])
```

11-1 linear for loop with if statement

add a new key to users dectionary. 
the values should be boolean.
the condition is: the age of users should be older than 30 

```python
users = {'names': ['John', 'Joe', 'Sara'], 'ages': [28, 50, 20]}
users['older_30'] = [True if age > 30 else False for age in users['ages']]
print(users)
```
